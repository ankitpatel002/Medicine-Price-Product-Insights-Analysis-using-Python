##  Missing Values Check

```python
df.isnull().sum()
```

```python
df['pack_size'] = df['pack_size'].fillna(round(df['pack_size'].mean())).astype(int)
```

```python
df.isnull().sum()
```

```python
df['pack_unit'] = df['pack_unit'].fillna(df['pack_unit'].mode()[0]).astype(str)
```

```python
df.isnull().sum()
```

```python
df['primary_strength'] = df['primary_strength'].fillna(df['primary_strength'].mode()[0]).astype(str)
```

```python
df.isnull().sum()
```

## KPI


```python
##Total Sales
total_sales = int(df['price_inr'].sum())

##Total Product
total_product = df['product_id'].nunique()

## Avg Pack Size
avg_pack_size = df['pack_size'].mean()


print(f"Total Sales: ${total_sales:,.0f}")
print(f"Total Product : {total_product:,}")
print(f"Avg Pack Size: {avg_pack_size:,.0f} Units")
```

```python
# Products with discount
discounted_count = df[df['is_discontinued'] == 'False']['product_id'].nunique()
print(f"Discounted Products: {discounted_count:,}")

qdiscounted_count = df[df['is_discontinued'] == 'True']['product_id'].nunique()
print(f"Discounted Products: {qdiscounted_count:,}")
```

## Visuals


```python
is_discontinued = df['is_discontinued'].value_counts()
plt.figure(figsize=(5,5))
is_discontinued.plot(
    kind = 'pie',
    autopct = '%1.1f%%',
    startangle = 140,
    colors = plt.cm.Pastel1.colors
)

plt.title('Product Distribution by discontinued')
plt.ylabel('')  
plt.tight_layout()
plt.show()
```

## Bar Chart — Top 10 Most Common Dosage Forms

```python
dosage_count = df['dosage_form'].value_counts().nlargest(10)
dosage_count.plot(kind = 'bar', color = 'darkgreen')
plt.title("Top 10 Most Common Dosage Forms")
plt.xlabel("Dosage Form")
plt.ylabel("No of Products")
plt.xticks(rotation=45)
plt.show()
```

## Histogram — Price Distribution


```python
sns.histplot(df['price_inr'], bins = 50, kde = True)
plt.title("Price Distribution (INR)")
plt.xlabel("Price")
plt.xlabel("Frequency")
plt.xlim(0,1000)
plt.show()
```

## Box Plot — Price by Dosage Form

```python
top_forms = df['dosage_form'].value_counts().nlargest(10).index
sns.boxplot(data = df[df['dosage_form'].isin(top_forms)], x = 'dosage_form',y = 'price_inr')
plt.title("Price Distrubution by Dosage Form")
plt.xticks(rotation = 45)
plt.ylim(0,10000)
plt.show()
```

## Scatter Plot — Price vs Number of Active Ingredients


```python
plt.scatter(df['num_active_ingredients'],df['price_inr'],alpha = 0.2)
plt.title("price vs. Number of active Ingredients")
plt.xlabel("Active Ingredients")
plt.xlabel("Price INR")
plt.ylim(0, 1000)
plt.show()
```

##  Heatmap — Correlation of Numeric Fields


```python
sns.heatmap(df[['price_inr', 'pack_size', 'num_active_ingredients']].corr(), annot=True, cmap='RdYlGn')
plt.title("Correlation Heatmap")
plt.show()
```

##  Count Plot — Top Therapeutic Classes


```python
top_classes = df['therapeutic_class'].value_counts().nlargest(10)
sns.barplot(x = top_classes.values, y=top_classes.index , palette ="mako")
plt.show()
```

## Donut Chart — Top 5 Pack Units

```python
pack_data = df['pack_unit'].value_counts().nlargest(4)
plt.pie(pack_data, labels=pack_data.index, autopct='%1.1f%%', startangle=140 )
plt.title("Top 5 Pack Units")
plt.show()
```
